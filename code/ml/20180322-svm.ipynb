```python
import numpy as np
import matplotlib.pyplot as plt

IMG_PATH = '../../static/images/ml-20180322-svm_{}.svg'
```

## 1. Figure 1

```python
np.random.seed(20)

# data set
X_0 = np.random.multivariate_normal([2,2], [[.4,.2],[.2,.4]], 8)
X_1 = np.random.multivariate_normal([6,3], [[.4,.2],[.2,.4]], 10)
plt.plot(X_0[:,0], X_0[:,1], 'bo', mec='k', ms=10)
plt.plot(X_1[:,0], X_1[:,1], 'ro', mec='k', ms=10)

# linear separation lines
x1 = np.arange(3,6,.5)
plt.plot(x1, -x1+6.5, 'k')
plt.text(5.6, 1, r'$\mathbf{w}^T\mathbf{x}+b=0$')
x2 = np.arange(3,5,.5)
plt.plot(x2, 2*x2-5, 'k')

plt.xticks([])
plt.yticks([])
plt.savefig(IMG_PATH.format(1))

plt.show()
```

## 2. Figure 2

```python
np.random.seed(20)

# data set
X_0 = np.random.multivariate_normal([2,2], [[.4,.2],[.2,.4]], 8)
X_1 = np.random.multivariate_normal([6,3], [[.4,.2],[.2,.4]], 10)
plt.plot(X_0[:,0], X_0[:,1], 'bo', mec='k', ms=10)
plt.plot(X_1[:,0], X_1[:,1], 'ro', mec='k', ms=10)

# linear separation lines
x1 = np.arange(2.6,3.2,.1)
plt.plot(x1, -7*x1+22, 'k--')
plt.text(2.5, -.7, r'$\mathbf{w}^T\mathbf{x}+b=-1$')
x2 = np.arange(4.6,5.3,.1)
plt.plot(x2, -7*x2+36.5, 'k--')
plt.text(5.4, -.5, r'$\mathbf{w}^T\mathbf{x}+b=1$')
x = np.arange(3.6,4.31,.1)
plt.plot(x, -7*x+29.3, 'k')
plt.text(4.4, -.9, r'$\mathbf{w}^T\mathbf{x}+b=0$')
plt.annotate('',xy=(3.13, .27*3.13-1), xytext=(5.2, .27*5.2-1), arrowprops=dict(arrowstyle='<->'))
plt.annotate('margin',xy=(4, .27*4-.1), xytext=(5.1, .27*5.1-.1), arrowprops=dict(arrowstyle='<->',color='r'))

plt.xticks([])
plt.yticks([])
plt.savefig(IMG_PATH.format(2))

plt.show()
```

## 3. Figure 3

```python
np.random.seed(20)

# data set
X_0 = np.random.multivariate_normal([2,2], [[.4,.2],[.2,.4]], 8)
X_1 = np.random.multivariate_normal([6,3], [[.4,.2],[.2,.4]], 10)
plt.plot(X_0[:,0], X_0[:,1], 'bo', mec='k', ms=10)
plt.plot(X_1[:,0], X_1[:,1], 'ro', mec='k', ms=10)

# linear separation lines
x1 = np.arange(2.6,3.2,.1)
plt.plot(x1, -7*x1+22, 'k--')
plt.text(2.5, -.7, r'$\mathbf{w}^T\mathbf{x}+b=-1$')
x2 = np.arange(4.6,5.3,.1)
plt.plot(x2, -7*x2+36.5, 'k--')
plt.text(5.4, -.5, r'$\mathbf{w}^T\mathbf{x}+b=1$')
x = np.arange(3.6,4.31,.1)
plt.plot(x, -7*x+29.3, 'k')
plt.text(4.4, -.9, r'$\mathbf{w}^T\mathbf{x}+b=0$')
plt.annotate('',xy=(3.13, .27*3.13-1), xytext=(5.2, .27*5.2-1), arrowprops=dict(arrowstyle='<->'))

# noise
e = np.array([[2.5,.5], [5, 4]])
plt.plot(e[0][0], e[0][1], 'ro', mec='k', ms=10)
plt.annotate(r'$\xi_i$',xy=(e[0][0]+.05, e[0][1]), xytext=(5.13, 1.1), arrowprops=dict(arrowstyle='<->',color='r'))
plt.plot(e[1][0], e[1][1], 'bo', mec='k', ms=10)
plt.annotate(r'$\xi_j$',xy=(e[1][0]-.05, e[1][1]), xytext=(2.45, 3.3), arrowprops=dict(arrowstyle='<->',color='b'))

plt.xticks([])
plt.yticks([])
plt.savefig(IMG_PATH.format(3))

plt.show()
```
