# Sigmoid Shape

```python
import numpy as np
import matplotlib.pyplot as plt

IMG_PATH = '../../static/images/ml-20180423-nn_{}.svg'
```

```python
np.random.seed(20)

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

x = np.arange(-6.0,6.0,.5)
plt.plot(x, sigmoid(x))

# plt.title('Sigmoid Function')
plt.xlabel('z')
plt.ylabel(r'$\sigma(z)$')
plt.grid()

plt.savefig(IMG_PATH.format('sigmoid_shape'))

plt.show()
```
